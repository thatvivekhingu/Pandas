```python
import pandas as pd
import matplotlib.pyplot as plt
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
import pandas as pd
import plotly.graph_objs as go
```

```python
df=pd.read_csv('directory.csv')
```

```python
df.head()
```

```python
df['Store Count'] = df.groupby('State/Province')['Store Number'].transform('count')
display(df.head())
```

```python
data = dict(type='choropleth',
locations = df['State/Province'],
locationmode = 'USA-states',
colorscale = 'Reds',
text = df['Brand'],
z = df['Store Count'],
colorbar = {'title':'Stores Count'})
layout =dict(title ='Starbucks locations in USA',geo = dict(scope='usa'))
choromap=go.Figure(data=[data],layout=layout)
iplot(choromap)
```
